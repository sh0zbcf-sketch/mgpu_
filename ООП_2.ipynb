```python
# Лабораторная работа 4.2 ООП. Наследование. Инкапсуляция task 27
# Выполнила: Шаповалова Софья
# Курс: ЦИБ-251
# =====================================================================
# Иерархия классов AudioContent (Базовый класс 1)
# =====================================================================
class AudioContent:
    """Базовый класс для музыкального контента."""

    def __init__(self, content_id: int, title: str): self.id = content_id; self.title = title
    def get_duration_in_seconds(self) -> int:
      return 0
    def __str__(self) -> str:
      return f"'{self.title}'"


class Track(AudioContent):
    """Дочерний класс. Музыкальный трек."""

    def __init__(self, content_id: int, title: str, artist: str, duration_seconds: int):
      super().__init__(content_id, title); self.performer = artist; self.duration_seconds = duration_seconds
    def get_duration_in_seconds(self) -> int:
      return self.duration_seconds
    def __str__(self) -> str:
      return f"🎵 Трек: {self.performer} — {super().__str__()}"


class Podcast(AudioContent):
    """Дочерний класс. Подкаст."""

    def __init__(self, content_id: int, title: str, author: str, clean_duration: int, intro_duration: int):
      super().__init__(content_id, title); self.author = author; self.clean_duration = clean_duration; self.intro_duration = intro_duration
    def get_duration_in_seconds(self) -> int:
      return self.clean_duration + self.intro_duration
    def __str__(self) -> str:
      return f"🎙️ Подкаст (Автор: {self.author}): {super().__str__()}"


# =====================================================================
# Иерархия классов Listener (Базовый класс 2)
# =====================================================================
class Listener:
    """Базовый класс для обычного слушателя."""

    def __init__(self, listener_id: int, username: str): self.id = listener_id; self.username = username
    def has_ads(self) -> bool:
      return True
    def __str__(self) -> str:
      return f"Пользователь {self.username} [Обычный]"


class PremiumListener(Listener):
    """Дочерний класс. Премиум-слушатель."""

    def __init__(self, listener_id: int, username: str):
      super().__init__(listener_id, username); self.no_ads = True
    def has_ads(self) -> bool:
      return False
    def __str__(self) -> str:
      return f"👑 Пользователь {self.username} [Premium, Без рекламы]"


# =====================================================================
# Класс-контейнер Playlist (Композиция)
# =====================================================================
class Playlist:
    """Класс-контейнер (Плейлист)."""

    def __init__(self, listener: Listener, playlist_name: str):
      self.listener = listener; self.name = playlist_name; self.content_list: List[AudioContent] = []
    def add_item(self, item: AudioContent) -> None:
      self.content_list.append(item)

    def get_total_duration(self) -> int:
        """Вычисляет общее время с учетом штрафа на рекламу."""
        total_seconds = sum(item.get_duration_in_seconds() for item in self.content_list)
        if self.listener.has_ads() and len(self.content_list) > 0:
            total_seconds += (len(self.content_list) // 2) * 30
        return total_seconds

    def print_playlist_info(self) -> None:
        """Выводит структурированный отчет по плейлисту."""
        print(f"--- Плейлист: '{self.name}' ---\nСлушатель: {self.listener}\nСодержимое:")
        for idx, item in enumerate(self.content_list, 1):
            print(f"  {idx}. {item} | Длительность: {item.get_duration_in_seconds()} сек.")
        total = self.get_total_duration()
        print(f"Итоговое время звучания: {total // 60} мин. {total % 60} сек.\n" + "-" * 45 + "\n")


# =====================================================================
# Демонстрация сценария
# =====================================================================
if __name__ == "__main__":
    print("=== ЗАПУСК МУЗЫКАЛЬНОГО СЕРВИСА ===\n")

    # 1. Создание клиентов
    free_user = Listener(listener_id=1, username="denis_rock")
    premium_user = PremiumListener(listener_id=2, username="anna_music")

    # 2. Создание объектов аудиоконтента
    track_1 = Track(101, "Smells Like Teen Spirit", "Nirvana", 301)
    track_2 = Track(102, "Numb", "Linkin Park", 185)
    podcast_1 = Podcast(201, "Как учить Python", "Хабр", 1200, 45)
    podcast_2 = Podcast(202, "История Рока", "Arzamas", 2400, 60)

    # 3. Наполнение плейлиста для обычного пользователя
    playlist_free = Playlist(free_user, "Мой Микс (Free)")
    for track in [track_1, track_2, podcast_1, podcast_2]: playlist_free.add_item(track)

    # 4. Наполнение плейлиста для Premium пользователя
    playlist_premium = Playlist(premium_user, "Премиум Медиатека")
    for track in [track_1, track_2, podcast_1, podcast_2]: playlist_premium.add_item(track)

    # Вывод отчетов в консоль
    playlist_free.print_playlist_info()
    playlist_premium.print_playlist_info()

```
