```python
# Задание task_03_03_05.
#
# Выполнил: Шаповалова С.С.
# Группа: ЦИБ-251

"""
Исправленная программа
"""

import random

def non_negatives(nums):
    i = 0
    while True:
        try:
            if nums[i] < 0:
                del nums[i]
            else:
                i += 1  # Сдвигаем индекс, только если элемент НЕ удален
        except IndexError:
            # Когда i выйдет за пределы уменьшившегося списка, цикл завершится
            break
    return nums

n = 10
nums = [round(random.uniform(-10, 10), 2) for i in range(n)]
print("Исходный список:", nums)

non_negatives(nums)
print("Результат:", nums)

```
