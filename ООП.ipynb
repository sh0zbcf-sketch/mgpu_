```python
#Лабораторная работа 4.1 ООП. Создание простых классов. Методы класса. task 27
#Выполнила: Шаповалова Софья
#Курс: ЦИБ-251

# =====================================================================
# Создание базового класса для основной сущности
# =====================================================================
class AudioContent:
    """Базовый класс для музыкального контента (треки, альбомы, подкасты)."""

    def __init__(self, content_id: int, title: str, artist: str, duration: int):
        self.id = content_id          # Указан в задании (id)
        self.title = title            # Название трека
        self.artist = artist          # Исполнитель
        self.duration = duration      # Длительность в секундах

    def display_info(self):
        """Выводит данные атрибутов в консоль."""
        print(f"ID Контента: {self.id}")
        print(f"Название: {self.title}")
        print(f"Исполнитель: {self.artist}")
        print(f"Длительность: {self.duration} сек.")

    # =====================================================================
    # Реализация «Vibe Code» (Читаемый вывод)
    # =====================================================================
    def __str__(self):
        """Возвращает красиво отформатированную строку с информацией о контенте."""
        minutes = self.duration // 60
        seconds = self.duration % 60
        return f"🎵 Аудио: {self.artist} — {self.title} | ID: {self.id} | Время: {minutes}:{seconds:02d}"


# =====================================================================
# Создание базового класса для второй сущности
# =====================================================================
class Listener:
    """Базовый класс для субъекта действия (Слушателя)."""

    def __init__(self, listener_id: int, username: str, subscription_type: str):
        self.id = listener_id                  # Указан в задании (id)
        self.username = username               # Имя пользователя
        self.subscription_type = subscription_type  # Тип подписки (Free / Premium)

    # =====================================================================
    # Реализация «Vibe Code» (Читаемый вывод)
    # =====================================================================
    def __str__(self):
        """Возвращает красиво отформатированную строку с информацией о слушателе."""
        return f"👤 Слушатель: {self.username} | ID: {self.id} | Подписка: [{self.subscription_type}]"


# =====================================================================
# Демонстрация работы программы
# =====================================================================
if __name__ == "__main__":
    print("--- Демонстрация работы музыкального сервиса ---\n")

    # Создание 3 объектов первого класса (AudioContent) с разными данными
    track_1 = AudioContent(content_id=101, title="Blinding Lights", artist="The Weeknd", duration=200)
    track_2 = AudioContent(content_id=102, title="Bohemian Rhapsody", artist="Queen", duration=355)
    track_3 = AudioContent(content_id=103, title="Shape of You", artist="Ed Sheeran", duration=233)

    # Создание 3 объектов второго класса (Listener)
    user_1 = Listener(listener_id=1, username="alex_rocker", subscription_type="Premium")
    user_2 = Listener(listener_id=2, username="maria_pop", subscription_type="Free")
    user_3 = Listener(listener_id=3, username="dj_ivan", subscription_type="Premium")

    # Вывод всех объектов в консоль с использованием print() для демонстрации __str__
    print("Список доступного аудиоконтента:")
    print(track_1)
    print(track_2)
    print(track_3)

    print("\nСписок зарегистрированных слушателей:")
    print(user_1)
    print(user_2)
    print(user_3)

    print("\n--- Демонстрация работы метода display_info() ---")
    track_1.display_info()
```
