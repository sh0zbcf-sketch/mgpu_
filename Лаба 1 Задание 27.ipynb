```python
# Задание Лабораторная работа 1, задание 27
#
# Выполнил: Шаповалова С.С
# Группа: ЦИБ-251
def calculate(expression):
    # 1. Сначала разбиваем строку по знаку ПЛЮС
    parts = expression.replace(" ", "").split('+')
    total_sum = 0

    for part in parts:
        # 2. Если внутри части есть УМНОЖЕНИЕ, считаем его
        if '*' in part:
            factors = part.split('*')
            product = 1
            for f in factors:
                product *= int(f)
            total_sum += product
        else:
            # 3. Если только число — просто прибавляем
            total_sum += int(part)

    return total_sum

# Проверка: 23 * 2 + 2 * 3 = 46 + 6 = 52
expr = input()
print(calculate(expr))
```
