```python
# Задание Лабораторная работа 1, задание 27
#
# Выполнил: Шаповалова С.С
# Группа: ЦИБ-251
def calculate(expression):
    # Убираем пробелы, чтобы они не мешали
    tokens = expression.replace(" ", "")

    nums = []   # Стек для чисел
    ops = []    # Стек для знаков (+ и *)

    # Задаем приоритеты операций
    priority = {'+': 1, '*': 2}

    for char in tokens:
        if char.isdigit():
            # Если это цифра, превращаем в число и кладем в стек
            nums.append(int(char))
        else:
            # Если это знак, проверяем приоритет того, что уже в стеке
            while ops and priority[ops[-1]] >= priority[char]:
                # Вытаскиваем данные для расчета
                op = ops.pop()
                b = nums.pop()
                a = nums.pop()

                # Расписываем расчет в несколько строк:
                if op == '+':
                    result = a + b
                else:
                    result = a * b

                # Кладем результат обратно в числа
                nums.append(result)

            # После очистки стека от сильных знаков, кладем текущий знак
            ops.append(char)

    # Когда строка кончилась, досчитываем то, что осталось в стеках
    while ops:
        op = ops.pop()
        b = nums.pop()
        a = nums.pop()

        if op == '+':
            result = a + b
        else:
            result = a * b

        nums.append(result)

    # Итоговый ответ — единственное число в стеке
    return nums[0]

stroka = input()
# Тест
print(calculate(stroka))
```
